```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
credits=pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#merge data frames and reassign it to movies
movies=movies.merge(credits,on= 'title')
```

```python
movies.head(1)
```

```python
#remove cols that are not necessary
#genres, id, keywords, title, overview, cast, crew   

movies=movies[['movie_id','title','overview','genres', 'keywords','cast','crew']]
```

```python
movies.head()
```

```python
#preprocessing
#I will create a new df which will have , movie_id, title and tags(consisting of overview merged with genres, keywords, cast ,crew)

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#Bring these columns in correct format
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(text):
    L = []
    counter=0
    for i in ast.literal_eval(text):
        if counter!=3:
            L.append(i['name']) 
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name']) 
            break
    return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# overview column is in string, convert it into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#remove spaces between words
## Most imporan thing for recommender systems
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies.head()
```

```python
#concatenate these lists and then convert it back to string
movies['tags']=movies['overview']+movies['cast']+movies['crew']+movies['keywords']+movies['genres']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
#covert into string 
new_df['tags']=new_df['tags'].apply(lambda x:" ". join(x))
```

```python
new_df.head()
```

```python
import nltk
# stem- it converts similar words to one kind
#eg- [love, loving, lovely] to [love, love, love]
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    res=[]
    for i in text.split():
        res.append(ps.stem(i))
        
    return " ".join(res)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#lower case conversion  // it is recommended
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
#Text vectorization- BAG OF WORDS TECHNIQUE, word2vec, tf idf etc
# I have to recommend a movie on the basis of similarity between tags of the dataset

#steps-  
#1 concatenate all the tags and then find the most common words
#2 for eg i got 5000 most common words they will become my headings of the columns
#3 Then I will start counting the freq of those words in the headings in each individual tags
#4 I'll get a table and each row will be my vector in 5000 dimensional space
```

```python
#We will use COUNTVECTORIZER class of scikit learn to get rid of meaningless words known as stop words

from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python

feature_names = cv.get_feature_names_out()
len(feature_names)
```

```python
new_df['tags'][0]
```

```python
# we'll cal cosine distance between vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]# sorting on the basis of 2nd column 
```

```python
# steps
# we'll sort the similarities in descending order and return top 5 movies
# for doing that we would reqire the index of the movie we are fetching to be able to compare it's features
# we can't sort aise hi coz then we'll lose the respective index value so to overcome that issue we will use ENUMERATE function 
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index] 
    movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]# sorting on the basis of 2nd column 
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
       
```

```python
recommend('Avatar')
```
